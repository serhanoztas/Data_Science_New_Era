```python
with engine.connect( ) as conn:
    df_sales = pd.read_sql (text("SELECT * FROM [dbo].[sales] "),conn)
    df_region = pd.read_sql (text("SELECT * FROM [dbo].[region_targets]"),conn)




```

```python
df_region['target_margin_pct'] = round(df_region['profit_target']/df_region['revenue_target']*100, 2)
df_region

```

```python
# df_sales.info()

df_merged = df_sales.merge(
    df_region,
    how='left',
    left_on='region',
    right_on='region',
    suffixes=('_sales','_region')
    )

df_merged.head()
```

```python
df_merged =df_merged.assign(
    revenue_per_unit=lambda d: d['total_revenue'] / d['units_sold'],
    margin_gap_vs_target=lambda d: d['profit_margin_pct'] - d['target_margin_pct'],
    days_to_ship=lambda d: d['ship_date'] - d['order_date']
)

df_merged.head()
```
