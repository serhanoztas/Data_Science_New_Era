SELECT 
    YEAR(Order_Date) AS order_year,
    MONTH(Order_Date) AS order_month,
    MONTHNAME(Order_Date) AS month_name,
    COUNT(*) AS order_count,
    SUM(Total_Revenue) AS total_revenue,
    SUM(Total_Profit) AS total_profit,
    AVG(Total_Revenue) AS avg_revenue,
    CASE 
        WHEN SUM(Total_Revenue) > AVG(SUM(Total_Revenue)) OVER (PARTITION BY MONTH(Order_Date)) 
            THEN 'above_monthly_avg'
        ELSE 'below_monthly_avg'
    END AS performance_vs_avg
FROM sales
GROUP BY YEAR(Order_Date), MONTH(Order_Date), MONTHNAME(Order_Date)
ORDER BY order_year, order_month;

```python
df_datediff['Order_Year'] = df_datediff['Order Date'].dt.year
df_datediff['Order_Month'] = df_datediff['Order Date'].dt.month
df_datediff['Order_MonthName'] = df_datediff['Order Date'].dt.month_name()
df_datediff

df_staging1 = df_datediff.groupby(['Order_Year','Order_Month','Order_MonthName']).agg(
    order_count = ('Order ID','count'),
    total_revenue = ('Total Revenue',('sum', 'mean')),
    total_profit = ('Total Profit','sum'),
    avg_revenue = ('Total Revenue','mean'),
).reset_index().round(2)

monthly_avg = df_datediff.groupby('Order_Month')['Total Revenue'].sum()
df_staging1['month_avg_revenue'] = df_staging1['Order_Month'].map(monthly_avg)

df_staging1

# total_revenue = (('Total Revenue','sum')>('monthly_avg'))
df_staging1['status'] = np.where(
    df_staging1['total_revenue'] > df_staging1['month_avg_revenue'],
    'above',
    'below'
)

df_staging1.head(20)
# 1197951.65
```
