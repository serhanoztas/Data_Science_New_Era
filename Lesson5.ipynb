```python
import pandas as pd
import numpy as np

filepath = r'C:\Users\serha\OneDrive\Desktop\Data_Science_New_Era\Data_Science_New_Era\dataset.csv'

df = pd.read_csv(filepath)
#df.info()

df = pd.read_csv(filepath, parse_dates=['Order Date','Ship Date'] )
df.info()

```

```python
df_avg = df.groupby('Item Type').agg(
    Avg_Total_Profit=('Total Profit','mean')
    )
df_avg

```

```python
# Order Counts by Itemtype
df_ordercounts = df.groupby('Item Type').agg(
    Ordercounts = ('Order ID','count')
)

df_ordercounts
```

```python
#df_avg
#df_ordercounts

df_merge = pd.merge(df_avg,df_ordercounts,on='Item Type')
df_merge
df_merge['Calculated_Total_Revenue'] = df_merge['Avg_Total_Profit']*df_merge['Ordercounts']
df_merge['Calculated_Total_Revenue']
```

```python
# sum of total rev, units, the number of orders by item type
# avg total porift and avg unit price by item type
# merge
# draw the values in graph

df_a = df.groupby('Item Type').agg(
    Total_Revenue=('Total Revenue','sum'),
    Total_Unit_Sold=('Units Sold','sum'),
    Order_Counts=('Order ID','count')
).reset_index()
df_a
df_a.columns = ['Item Type','Total Revenue', 'Total Units Sold', 'Order Count']
df_a

df_b = df.groupby('Item Type').agg(
    AVG_Profit=('Total Profit','mean'),
    AVG_Unit_price=('Unit Price','mean')
).reset_index()

df_b.columns = ['Item Type','Avg Profit','Avg Unit Price']
df_b

df_a_b_merge = pd.merge(df_a,df_b, on='Item Type')
df_a_b_merge
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2)
fig.suptitle('sales', fontsize=12)

# total revenue
axes[0].bar(df_a_b_merge['Item Type'], df_a_b_merge['Total Revenue']/1000000, color='blue')
axes[0].set_xlabel('total revenue in mln', fontsize=8)
axes[0].set_title('profitability', fontsize=8)
axes[0].grid(axis='x')

# total revenue
axes[1].bar(df_a_b_merge['Item Type'], df_a_b_merge['Total Revenue']/1000000, color='blue')
axes[0].set_xlabel('total revenue in mln', fontsize=8)
axes[0].set_title('profitability', fontsize=8)
axes[0].grid(axis='x')

plt.show()
```
