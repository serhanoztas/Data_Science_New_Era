1. Create array: [1200, 1800, 950, 2200, 2500, 3100, 2800]
2. Calculate total weekly sales
3. Find the average daily sales
4. Find days with above-average sales
5. Calculate percentage growth from day 1 to day 7


```python
weeklysales = np.array([1200, 1800, 950, 2200, 2500, 3100, 2800])
print (weeklysales)
print(np.sum(weeklysales))
print(np.mean(weeklysales))
print(weeklysales[weeklysales>np.mean(weeklysales)])
day1 = weeklysales[0]
day7 = weeklysales[-1]
print((day7-day1)/day1*100)
```

You have quarterly sales data for 4 products:

           Q1    Q2    Q3    Q4
Product A: 1500, 1800, 2100, 2400
Product B: 900,  1100, 1000, 1200
Product C: 2000, 2200, 1900, 2500
Product D 1200, 1300, 1400, 1100

Tasks:
1. Create the 2D array
2. Calculate total annual sales per product
3. Find the best quarter for each product
4. Calculate average quarterly sales across all products
5. Find which product had the highest single-quarter sales
6. Get all sales figures above 2000


```python
import numpy as np

# sales = np.array([
#     [1500, 1800, 2100, 2400],  # Product A
#     [ 900, 1100, 1000, 1200],  # Product B
#     [2000, 2200, 1900, 2500],  # Product C
#     [1200, 1300, 1400, 1100]   # Product D
# ])
print("------------------")

ProductA = np.array([1500, 1800, 2100, 2400])
ProductB = np.array([ 900, 1100, 1000, 1200])
ProductC = np.array([2000, 2200, 1900, 2500])
ProductD = np.array([1200, 1300, 1400, 1100])
print("------------------")
sales = np.array([
ProductA,
ProductB,
ProductC,
ProductD
])

print(sales)
print("------------------")

sumsales = np.sum([
ProductA,
ProductB,
ProductC,
ProductD
])

print(sumsales)
print("------------------")

maxsales = np.array([
np.max(ProductA),
np.max(ProductB),
np.max(ProductC),
np.max(ProductD)
])

print(maxsales)
print("------------------")

print(sales[sales>2000])
```
