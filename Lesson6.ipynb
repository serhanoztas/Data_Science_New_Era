```python
import numpy as np
import pandas as pd

filepath = r'C:\Users\serha\OneDrive\Desktop\Data_Science_New_Era\Data_Science_New_Era\dataset.csv'
df = pd.read_csv(filepath, parse_dates=['Order Date','Ship Date']).reset_index()
df
```

```python
df_unitsold = df['Units Sold']
print(type(df_unitsold))
# for item in dir(df_unitsold):
#     print(item)

units_sold = df_unitsold.to_numpy(dtype=int)
print(units_sold)
```

```python
df_unit_sold = df['Units Sold'].to_numpy(dtype=float)
df_unit_price = df['Unit Price'].to_numpy(dtype=float)
df_unit_cost = df['Unit Cost'].to_numpy(dtype=float)
df_total_cost = df['Total Cost'].to_numpy(dtype=float)
df_total_revenue = df['Total Revenue'].to_numpy(dtype=float)
df_total_profit = df['Total Profit'].to_numpy(dtype=float)







```

```python
# select total_profit/units_sold as profit_per_unit from dataset where units_sold > 0


df_calc1 = np.divide(df_total_profit,df_unit_sold,out=np.full_like(df_total_profit, np.nan), where=df_unit_sold!=0)
df_calc1
```

```python
# select total_profit/total_revenu as profit_margin where total_rev > 0

df_calc2 = np.divide(df_total_profit,df_total_revenue, out=np.full_like(df_total_profit, np.nan), where=df_total_revenue!=0)
df_calc2
```

```python
# select (unit_rpie - unit_cost) / unit_price as ration where unit_priec > 0
df_calc3 = np.divide(np.subtract(df_unit_price,df_unit_cost),df_unit_cost, out=np.full_like(df_unit_price,np.nan), where=df_unit_price!=0)
df_calc3
```

```python
# profit per unit, profit margin, ratio
df['profit per unit'] = df_calc1
df['profit margin'] = df_calc2
df['ratio'] = df_calc3

df_limited = df[['Units Sold','Unit Price','Unit Cost','Total Revenue','Total Cost','Total Profit','profit per unit','profit margin','ratio']]
df_limited.info()


```

```python
# select percentile_count(0.75) within group (order by total_profit) from dataaset.
counter = 10



profit_75 = np.nanpercentile(df_total_profit, 75)
revenue_median = np.nanpercentile(df_total_revenue, 50)
unit_80 = np.nanpercentile(df_unit_sold, 80)

mask = (
    (df_total_profit >= profit_75) &
    (df_total_revenue >= revenue_median) &
    (units_sold>=unit_80)
)

# select * from dataset wehre total_proft >= [75th]and total_Rev>50
df['new_col'] = np.where(
    mask, 'high', 'standard'
)

df_filtered =  df.loc[df['new_col'] == 'high']
df_filtered
```
