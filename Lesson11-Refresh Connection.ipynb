# SELECT
#     CASE
#         WHEN total_revenue < 10000  THEN 'Small'
#         WHEN total_revenue < 100000 THEN 'Medium'
#         WHEN total_revenue < 500000 THEN 'Large'
#         ELSE 'XLarge'
#     END AS order_size,
#     AVG(profit_margin_pct) AS avg_margin,
#     COUNT(*) AS order_count
# FROM sales
# GROUP BY order_size
# ORDER BY order_count DESC;

```python
conditions = [
    df_sales['total_revenue'] < 10000 ,
    df_sales['total_revenue'] < 100000,
    df_sales['total_revenue'] < 500000
]
choices = ['small','nedium','large']
default= 'xLarge'

df_sales['order_size'] = np.select(conditions, choices, default=default)
df_sales.head().reset_index()

df_sales_ordersize = df_sales.groupby('order_size').agg(
    avg_margin=('profit_margin_pct','mean'),
    order_count=('order_id','count'),
    total_rev_sum=('total_revenue','sum')
).sort_values('order_count', ascending=False)
    
df_sales_ordersize.reset_index()
```
