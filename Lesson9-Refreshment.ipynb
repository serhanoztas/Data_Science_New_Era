```python
import pandas as pd
import numpy as np
```

```python
filepath = r'C:\Users\serha\OneDrive\Desktop\Data_Science_New_Era\Data_Science_New_Era\dataset.csv'
df = pd.read_csv(filepath, parse_dates=['Order Date','Ship Date'])
#df.info()

df[['Region','Order Date','Total Profit']]


```

```python
df_totprofit = df[df['Total Profit']>500000]
df_totprofit
```

```python
#Total profit > 50000, Region = Europe

df_mult = df[(df['Total Profit']>50000) & (df['Region']=='Europe')]
df_mult
```

```python
# WHERE Item Type in Beverages, Snacks, Baby Food

df_mult2 = df[df['Item Type'].isin (['Baby Food','Beverages','Snacks'])]
df_mult2
```

```python
# Sales Channel = online, Order Priority in h,c, tot rev>100000

df_mult3 =  df[(df['Sales Channel']=='Online') 
               & (df['Order Priority'].isin (['H','C']))
                & (df['Total Revenue']>100000) ]
df_mult3
```

```python
#Sum Total Revenue, Group by Region

df_sumtotrevregion = df.groupby(['Region']).agg(
    Total_Revenue_region=('Total Revenue','sum')
).reset_index()
#print((df_sumtotrevregion['Total_Revenue_region'].astype(int)))
df_sumtotrevregion['Total_Revenue_region']=df_sumtotrevregion['Total_Revenue_region'].astype(int)
df_sumtotrevregion
```

```python
# count order id, sum total profit,  avg unit price, group by item type

df_mult_groupby = df.groupby(['Item Type']).agg(
    Count_Order_Id = ('Order ID','count'),
    Sum_Total_Profit = ('Total Profit','sum'),
    AVG_Unit_Price = ('Unit Price','mean')
).reset_index()

df_mult_groupby['Sum_Total_Profit']=df_mult_groupby['Sum_Total_Profit'].astype(int)
df_mult_groupby
```

```python
# Year, Month, Day from Orderdate

df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Day_Name'] = df['Order Date'].dt.day_name()
df['Shipping Days'] = (df['Ship Date']-df['Order Date']).dt.days  # 2024-31-12 - 2024-20-12 = 11

df
```

```python
# sum tot rev group by year, month

df_group_by_date = df.groupby(['Order Year','Order Month']).agg(
    sum_total_revenue = ('Total Revenue','sum')
)

df_group_by_date.sort_values(['Order Year','Order Month'],ascending=False).reset_index()
```
